## Local Metrics

```python
results = load_model('californiaHouseCR_results')
```

```python
# Local Metrics !
ce = np.array(results.dist_local)
inlier = np.mean(results.isolation.predict(ce) == 1)

t = 0
for s in results.S_star_local:
    t += len(s)
sparsity = t/len(results.S_star_local)
```

```python
print('LOCAL Accuracy = {} -- Coverage = {} - Inlier = {} - Sparsity = {}'.format(results.accuracy_local,
                                                                  results.coverage_local, 
                                                                  inlier, sparsity))
```
